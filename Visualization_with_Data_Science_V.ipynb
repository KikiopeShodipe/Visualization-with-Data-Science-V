```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read the data
df=pd.read_csv("/content/USA_Housing.csv")
df.head()
```

```python
#check for null values
df.isnull().sum()
```

#Findings
* There are no null values

```python
#dataset structures
df.info()
```

#Findings
* The dataset has 5000 entries with 7 columns
* 6 numerical columns and 1 descriptive columns

```python
#drop the descriptive column

df.drop("Address",axis=1,inplace=True)
```

```python
#Basic Scatter Plot: Area Income vs Price

plt.figure(figsize=(6, 4))

plt.scatter(df['Avg. Area Income'], df['Price'])

plt.title("1. Income vs House Price")

plt.xlabel("Avg. Area Income")

plt.ylabel("Price")

plt.show()
```

```python
plt.figure(figsize=(6, 4))

plt.scatter(df['Avg. Area House Age'], df['Price'], c=df['Area Population'], cmap='coolwarm', alpha=0.6)

plt.colorbar(label='Population')

plt.title("2. House Age vs Price (Colored by Population)")

plt.xlabel("Avg. Area House Age")

plt.ylabel("Price")

plt.show()
```
